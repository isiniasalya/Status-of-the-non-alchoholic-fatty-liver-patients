```python
ch.rename(columns={'Unnamed: 0':'Un'},inplace=True)
```

```python
ch.head()
```

```python
ch = ch.drop(columns=["Un"])
ch.head()
```

```python
import pandas as pd
```

```python
hdl_data = ch[ch['test'] == 'hdl']
chol_data = ch[ch['test'] == 'chol']
```

```python
def calculate_averages(data):
    pre_data = data[data['days'] < 0]  # Data before index date (days < 0)
    post_data = data[data['days'] >= 0]  # Data after index date (days >= 0)
    avg_pre = pre_data['value'].mean()
    avg_post = post_data['value'].mean()
    avg_total = data['value'].mean()
    return pd.Series([avg_pre, avg_post, avg_total], index=['avg_pre', 'avg_post', 'avg_total'])
```

```python
hdl_averages = hdl_data.groupby('id').apply(calculate_averages).reset_index()
chol_averages = chol_data.groupby('id').apply(calculate_averages).reset_index()
```

```python
hdl_averages
```

```python
merged_ch = pd.merge(hdl_averages, chol_averages, on='id')
```

```python
merged_ch.columns = ['id', 'avg_pre_hdl', 'avg_post_hdl', 'avg_total_hdl', 'avg_pre_chol', 'avg_post_chol', 'avg_total_chol']
```
